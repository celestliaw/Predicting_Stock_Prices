## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

## Import Data

```python
from yahoofinancials import YahooFinancials

stocks= ['MSFT', 'SPY','GOOG','ADBE','NKE','TSLA',"FB","DELL","HPQ","AMZN"]
```

```python
#original data before cleaning
yahoo_financials = YahooFinancials('WFC')
print(yahoo_financials.get_historical_price_data("2018-07-10", "2018-08-10", "monthly"))
```

## Data Extraction + Cleaning

```python
df=[]

    
for item in stocks:
    yahoo_financials = YahooFinancials(item)
    data = yahoo_financials.get_historical_price_data("2020-02-01", "2022-02-01", "monthly")
    df.append(pd.DataFrame(data[item]['prices']))
    df[-1]['name'] = item
    
    for key in data[item]['prices']:
        #print(key)
        continue
## delete the first coloumn , clean data
dataframe=pd.concat(df)
dataframe=dataframe.drop(columns=['date'])
display(dataframe)
```

## Importing economic variables

```python
#acquire retailsales data
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=RETAIL_SALES&apikey=4NABZ8IHZ6Q3TQ0H'
r = requests.get(url)
retailsales = r.json()


#appending retailsales data to dataframe
asd = [ ]
for x in range(0,25):
    asd.append(retailsales['data'][x]['value'])

print(asd[::-1])
newCol = pd.DataFrame(asd[::-1])
dataframe['RetailSales'] = newCol
dataframe.head(24)
```

```python
#acquire unemployment data
url = 'https://www.alphavantage.co/query?function=UNEMPLOYMENT&apikey=4NABZ8IHZ6Q3TQ0H'
r = requests.get(url)
unemployment = r.json()

#appending unemployment data to dataframe
asd = [ ]
for x in range(0,25):
    asd.append(unemployment['data'][x]['value'])

print(asd[::-1])
newCol = pd.DataFrame(asd[::-1])
dataframe['Unemployment'] = newCol
dataframe.head(25)
```

```python
#acquire consumer sentiment data
url = 'https://www.alphavantage.co/query?function=CONSUMER_SENTIMENT&apikey=4NABZ8IHZ6Q3TQ0H'
r = requests.get(url)
csentiment = r.json()

asd = [ ]
for x in range(0,25):
    asd.append(csentiment['data'][x]['value'])

print(asd[::-1])
newCol = pd.DataFrame(asd[::-1])
dataframe['CSentiment'] = newCol
dataframe.head(24)
```

```python
#acquire inflation expectation data
url = 'https://www.alphavantage.co/query?function=INFLATION_EXPECTATION&apikey=4NABZ8IHZ6Q3TQ0H'
r = requests.get(url)
inflationexp= r.json()

asd = [ ]
for x in range(0,25):
    asd.append(inflationexp['data'][x]['value'])

print(asd[::-1])
newCol = pd.DataFrame(asd[::-1]) #changed the order of formatted_date to match previous data
dataframe['InflationExp'] = newCol
dataframe.head(24)
```

```python
#acquire federal funds data
URL='https://www.alphavantage.co/query?function=FEDERAL_FUNDS_RATE&interval=monthly&apikey=4NABZ8IHZ6Q3TQ0H'

r = requests.get(url)
federalfunds= r.json()

asd = [ ]
for x in range(0,25):
    asd.append(federalfunds['data'][x]['value'])

print(asd[::-1])
newCol = pd.DataFrame(asd[::-1]) #changed the order to match previous data
dataframe['FederalFunds'] = newCol
dataframe.head(24)
```

## Creating new columns

```python
#creating new column 'change'
dataframe['Change'] = dataframe['close'] - dataframe['open']
               
dataframe
```

```python
#creating new column 'change_class'
def fun(num):
    if num <0:
        return "down"
    if num>0:
        return "up"
    if num==0:
        return "same"
    
dataframe['change_class'] = dataframe['Change'].apply(fun)
dataframe
```

## Changing variable type

```python
dataframe['RetailSales']=dataframe['RetailSales'].astype(int)  # changed from string to respective
dataframe['Unemployment']=dataframe['Unemployment'].astype(float)
dataframe['CSentiment']=dataframe['CSentiment'].astype(float)
dataframe['InflationExp']=dataframe['InflationExp'].astype(float)
dataframe['FederalFunds']=dataframe['FederalFunds'].astype(float)
dataframe.reset_index(drop=True)
```
